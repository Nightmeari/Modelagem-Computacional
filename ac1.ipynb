Desenvolva um programa bem estruturado para calcular a expansão em série
de Taylor da função cosseno. A função deve ter as seguintes características:
    • Permitir que o usuário especifique valores para as seguintes entradas;
        § Quantidade de termos que o usuário deseja utilizar;
        § Definir o ponto inicial (xi) e o ponto a ser previsto (xi+1)
    • Como saída o seu programa deverá mostrar as seguintes informações:
        o Devolver a estimativa de cos (x), e o erro relativo verdadeiro (que
        você pode calcular com base função cosseno).
        o Mostrar os pontos xi , xi+1 e a solução exata graficamente.


```python
from numpy import cos, linspace
from matplotlib.pyplot import plot, scatter, grid, legend, xlabel, ylabel, show

x0 = float(input("Digite o valor do x inicial: "))
x1 = float(input("Digite o valor do x final: "))
n  = int(input("Digite o número de termos da Série de Taylor: "))

f = lambda x: cos(x)

soma = 0

for k in range(n):
    df = cos(x0 + k * 3.141592653589793/2) 
    termo = df * (x1 - x0)**k
    fat = 1
    for i in range(1, k+1):
        fat *= i
    soma += termo / fat
aprox = soma

f_x0 = f(x0)
f_x1 = f(x1)

erro_rel = abs((f_x1 - aprox) / f_x1) * 100

print(f"O valor da função no ponto inicial = {x0} é: {f_x0}")
print(f"O valor da função no ponto final = {x1} é: {f_x1}")
print(f"O valor aproximado de ordem {n} é: {aprox}")
print(f"O erro relativo verdadeiro é: {erro_rel}%")

xs = linspace(x0 - 1, x1 + 1, 200)
ys = f(xs)

plot(xs, ys, label="f(x) = cos(x)")
scatter(x0, f_x0, label="ponto inicial")
scatter(x1, f_x1, label="ponto final")
scatter(x1, aprox, label=f"Taylor ordem {n}")

grid(True)
legend()
xlabel("x")
ylabel("f(x)")
show()
```
