Considere a função f(x) = x3-2x+4 no intervalo [-2,2] com h = 0,25.
Use aproximações por diferenças finitas progressivas, regressivas e centradas
para a primeira e segunda derivadas de modo a ilustrar graficamente qual
aproximação é mais acurada. Faça o gráfico de todas as três aproximações por
diferenças finitas da primeira derivada com a derivada teórica e faça o mesmo
também para a segunda derivada.

```python
from numpy import arange
import matplotlib.pyplot as plt

f = lambda x: x**3 - 2*x + 4
df = lambda x: 3*x**2 - 2
df2 = lambda x: 6*x

h = 0.25
x = arange(-2, 2+h, h)
y = f(x)

df_prog = []
df_reg = []
df_cent = []

for i in range(len(x)-1): # - 1 pq a prog nn pega o ultimo
    df_prog.append((y[i+1]- y[i])/h)

for i in range(1, len(x)): # - 1 pq a prog nn pega o ultimo
    df_reg.append((y[i]- y[i-1])/h)

for i in range(1, len(x)-1):
    df_cent.append((y[i+1] - y[i-1]) / (2*h))
x_prog = x[:-1]
x_reg = x[1:]
x_cent = x[1:-1]

df_teor=df(x)

plt.figure(figsize=(8,5))

plt.plot(x, df_teor, label='Teórica')
plt.plot(x_prog, df_prog, '--', label='Progressiva')
plt.plot(x_reg, df_reg, '--', label='Regressiva')
plt.plot(x_cent, df_cent, '--', label='Centrada')

plt.legend()
plt.title('Comparação - Primeira Derivada')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid()
plt.show()

df2_cent = []

for i in range(1, len(x)-1):
    df2_cent.append((y[i+1] - 2*y[i] + y[i-1]) / h**2)

x_cent2 = x[1:-1]
df2_teor = df2(x)

plt.figure(figsize=(8,5))

plt.plot(x, df2_teor, label='Teórica')
plt.plot(x_cent2, df2_cent, '--', label='Centrada')

plt.legend()
plt.title('Comparação - Segunda Derivada')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.grid()
plt.show()
```
