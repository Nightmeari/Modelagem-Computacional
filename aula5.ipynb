```python
import numpy as np
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel
from scipy.optimize import minimize

f = lambda V: 0.01*0.5*V**2 + (0.95/0.5)*((15000/V)**2)

V = np.linspace(100, 1200, 1000)

resultado = minimize(f, 400)

V_min = resultado.x[0]
D_min = f(V_min)

plot(V, f(V), label="Arrasto")
plot(V_min, D_min, "ro", label="Mínimo")
xlabel("Velocidade")
ylabel("Arrasto")
legend()
grid()

print(f"Velocidade ótima: {V_min}")
print(f"Arrasto mínimo: {D_min}")
```

```python
import numpy as np
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel
from scipy.optimize import minimize

f = lambda x: -4 + 1.8*x**2 + 1.2 *x**3 - 0.3*x**4
x = np.linspace(-50, 50, 1000)


metodo = int(input("Escolha o método: 1-Newton-Raphson, 2 - Bissecção e 3 - Minimize: "))


if metodo == 1:
    df = lambda x: 3.6*x + 3.6*x**2 - 1.2*x**3
    d2f = lambda x: 3.6 + 7.2*x - 3.6*x**2
    
    xi = float(input("Chute inicial: "))
    
    for i in range(10):
        xi = xi - df(xi)/d2f(xi)
        print(f"Iteração: {i+1}: x = {xi}")
    print(f"Newton:\n{xi}, {f(xi)}")
elif metodo == 2:
    xl = float(input("Limite inferior: ")) 
    xu = float(input("Limite superior: "))
    df = lambda x: 3.6*x + 3.6*x**2 - 1.2*x**3

    if df(xl)*df(xu) > 0:
        print("Intervalo inválido — não garante raiz de f'(x)")
        exit()

    erro = 100
    xaprox_ant = 0
    iter = 0
    while erro > 0.1:
        xaprox = (xl + xu)/2
        if iter > 0:
            erro = abs((xaprox - xaprox_ant)/xaprox)*100
        if df(xl)*df(xaprox)<0:
            xu = xaprox
        else:
            xl = xaprox
        xaprox_ant = xaprox
        iter += 1
    print(f"Bissecção:\n {xaprox}, {f(xaprox)}")
elif metodo == 3:
    g = lambda x: -f(x)
    resultado = minimize(g, -20)
    X_max = resultado.x[0]
    Y_max = f(X_max)
    print(f"Minimize:\n{X_max}, {Y_max}")

```
