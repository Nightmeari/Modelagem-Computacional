Utilize as expressões de derivação numérica para encontrar o valor da derivada da função a seguir em x=[1,3,5,7]
Faça um segundo teste reduzindo o valor de ∆𝑥 e discuta

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 3, 5, 7]
x2 = [1, 2, 3, 4, 5, 6, 7]
y = np.sin(0.5*np.sqrt(x))/x
y2 = np.sin(0.5*np.sqrt(x2))/x2

print(x)
print(y)
print(x2)
print(y2)

dx = np.gradient(y, x)
dx2 = np.gradient(y2, x2)

plt.plot(x, y, label="F(x)")
plt.plot(x2, y2, label="F(x^2)")
plt.plot(x, dx, 'ro', label='dy/dx com dx=2')
plt.plot(x2, dx2, 'go', label='dy/dx com dx=1')


plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
