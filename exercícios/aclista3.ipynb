```python
#1A -  Falsa Posição
import numpy as np

# Constantes
g = 9.81
L = 4
t = 2.5
v_desejado = 5

f = lambda H: np.sqrt(2*g*H) * np.tanh((np.sqrt(2*g*H)/(2*L))*t) - v_desejado

x0 = 0
xi = 2

erro = 100
xaprox_ant = 0
iter = 0

while erro > 1:
    
    xaprox = xi - (f(xi)*(x0 - xi)) / (f(x0) - f(xi)) # Falsa posição
    
    if iter > 0:
        erro = abs((xaprox - xaprox_ant)/xaprox)*100
    
    if f(x0)*f(xaprox) < 0:
        xi = xaprox
    else:
        x0 = xaprox
    
    xaprox_ant = xaprox
    iter += 1
    
    print(f"Iteração {iter}: Aproximação = {xaprox:.4f}, erro = {erro:.2f}%")

print("\nAltura encontrada H =", xaprox)
```

```python
#1B - Bissecção
import numpy as np

# Constantes
g = 9.81
L = 4
t = 2.5
v_desejado = 5

f = lambda H: np.sqrt(2*g*H) * np.tanh((np.sqrt(2*g*H)/(2*L))*t) - v_desejado

x0 = 0
xi = 2

erro = 100
xaprox_ant = 0
iter = 0

while erro > 1:
    
    xaprox = (x0 + xi)/2 # Bissecção
    
    if iter > 0:
        erro = abs((xaprox - xaprox_ant)/xaprox)*100
    
    if f(x0)*f(xaprox) < 0:
        xi = xaprox
    else:
        x0 = xaprox
    
    xaprox_ant = xaprox
    iter += 1
    
    print(f"Iteração {iter}: Aproximação = {xaprox:.4f}, erro = {erro:.2f}%")

print("\nAltura encontrada H =", xaprox)
```

Qual método foi mais eficiente? Por quê?
O método da falsa posição foi mais eficiente porque ele usa a informação da função, não só do intervalo. o método da bissecção ignora o formato da função, converge mais devagar e reduz o erro de forma linear, já a falsa posição chega mais perto da riaz com menos iterações

```python
#questão 5 b
import numpy as np

f = lambda x: 3 + 6*x + 5*x**2 + 3*x**3 + 4*x**4

def df(x):
    h = 0.0001
    return (f(x+h) - f(x-h))/(2*h)

def df2(x):
    h = 0.0001
    return (f(x+h) - 2*f(x) + f(x-h))/(h**2)

xi = 1

erro = 100
xant = 1
iter = 0

while erro > 1:
    
    xnovo = xi - df(xi)/df2(xi)
    
    if iter >= 0:
        erro = abs((xnovo - xant)/xnovo)*100
    
    xant = xnovo
    xi = xnovo
    iter += 1
    
    print(f"Iteração {iter}: x = {xnovo:.5f}, erro = {erro:.2f}%")

print("\nPonto de mínimo ≈", xnovo)
print("f(x) mínimo =", f(xnovo))
```
