```python
import numpy as np
#constantes:
g = 9.81
L = 4
t = 2.5
v_desejado = 5

f = lambda H: np.sqrt(2*g*H) * np.tanh(np.sqrt(2*g*H)/(2*L)*t) - v_desejado
xl = 0
xu = 2
#Falsa Pos:
xaprox_ant = 0
erro = 100
iter = 0

if f(xl)*f(xu) > 0:
        print("Erro: não há troca de sinal no intervalo")
else:
    while erro > 1:
        xaprox = xu - (f(xu)*(xl-xu)) / (f(xl) - f(xu))
        if iter > 0:
            erro = abs((xaprox - xaprox_ant)/xaprox)*100
        if f(xl)*f(xaprox) < 0:
            xu = xaprox
        else:
            xl = xaprox
        xaprox_ant = xaprox
        iter += 1
        print(f"Iteração: {iter} -> {xaprox}, {f(xaprox)}")
    print(f"Resultado:\n{xaprox}, {f(xaprox)}")
```

```python
import numpy as np

# constantes
g = 9.81
c = 15
t = 10

f = lambda m: (g*m)/c * (1 - np.e**(-c*t/m)) - 36

xl = 1
xu = 100
erro = 100
xaprox_ant = 0
iter = 0

fl = f(xl)
fu = f(xu)

if fl * fu > 0:
    print("Erro: Nesse intervalo não tem troca de sinal.")
else:
    while erro > 0.1:
        xaprox = xu - (fu*(xl-xu))/(fl-fu)
        fa = f(xaprox)

        if iter > 0:
            erro = abs((xaprox - xaprox_ant)/xaprox)*100

        if fl * fa < 0:
            xu = xaprox
            fu = fa
        else:
            xl = xaprox
            fl = fa

        xaprox_ant = xaprox
        iter += 1

        print(f"Iteração {iter}: {xaprox:.4f}, {fa:.4f}")

    print(f"FINAL\nIteração {iter}: {xaprox:.4f}, {fa:.4f}")
```

```python
import numpy as np

f = lambda x: 3 + 6*x + 5*x**2 + 3*x**3 + 4*x**4
df = lambda x: 6 + 10*x + 9*x**2 + 16*x**3
df2 = lambda x: 10 + 18*x + 48*x**2

xi = -1
erro = 100
iter = 0
if df2(xi) == 0:
    print("ERRO.")
else:
    while erro >= 1:
        fi = df(xi)
        f2i = df2(xi)
        xnovo = xi - fi/f2i
        if iter > 0:
            erro = abs((xnovo - xi)/xnovo)*100
            print(f"Iteração {iter+1}: x = {xnovo:.4f}, f(x) = {f(xnovo):.4f}, erro = {erro:.4f}")
        else:
            print(f"Iteração {iter+1}: x = {xnovo:.4f}")
        xi = xnovo
        iter +=1
```

```python
import numpy as np

f = lambda x: 3 + 6*x + 5*x**2 + 3*x**3 + 4*x**4

xi = 1
h = 0.01
erro = 100
iter = 0

while erro >= 1:
    
    f_xph = f(xi + h)
    f_xmh = f(xi - h)
    f_x = f(xi)
    
    # derivada primeira (diferença centrada)
    f1 = (f_xph - f_xmh) / (2*h)
    
    # derivada segunda (diferença centrada)
    f2 = (f_xph - 2*f_x + f_xmh) / (h**2)
    
    xnovo = xi - f1/f2
    
    if iter > 0:
        erro = abs((xnovo - xi)/xnovo)*100
    
    print(f"Iteração {iter+1}: x = {xnovo:.4f}, erro = {erro:.4f}")
    
    xi = xnovo
    iter += 1
```
