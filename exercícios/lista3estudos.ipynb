```python
import numpy as np
#constantes:
g = 9.81
L = 4
t = 2.5
v_desejado = 5

f = lambda H: np.sqrt(2*g*H) * np.tanh(np.sqrt(2*g*H)/(2*L)*t) - v_desejado
xl = 0
xu = 2
#Falsa Pos:
xaprox_ant = 0
erro = 100
iter = 0

if f(xl)*f(xu) > 0:
        print("Erro: não há troca de sinal no intervalo")
else:
    while erro > 1:
        xaprox = xu - (f(xu)*(xl-xu)) / (f(xl) - f(xu))
        if iter > 0:
            erro = abs((xaprox - xaprox_ant)/xaprox)*100
        if f(xl)*f(xaprox) < 0:
            xu = xaprox
        else:
            xl = xaprox
        xaprox_ant = xaprox
        iter += 1
        print(f"Iteração: {iter} -> {xaprox}, {f(xaprox)}")
    print(f"Resultado:\n{xaprox}, {f(xaprox)}")
```

```python
import numpy as np
#constantes:
g = 9.81
L = 4
t = 2.5
v_desejado = 5

f = lambda H: np.sqrt(2*g*H) * np.tanh(np.sqrt(2*g*H)/(2*L)*t) - v_desejado
xl = 0
xu = 2
#Bissecção:
xaprox_ant = 0
erro = 100
iter = 0

if f(xl)*f(xu) > 0:
        print("Erro: não há troca de sinal no intervalo")
else:
    while erro > 1:
        xaprox = (xl+xu)/2
        if iter > 0:
            erro = abs((xaprox - xaprox_ant)/xaprox)*100
        if f(xl)*f(xaprox) < 0:
            xu = xaprox
        else:
            xl = xaprox
        xaprox_ant = xaprox
        iter += 1
        print(f"Iteração: {iter} -> {xaprox}, {f(xaprox)}")
    print(f"Resultado:\n{xaprox}, {f(xaprox)}")
```
