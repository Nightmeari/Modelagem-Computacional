Use a série de Taylor para fazer uma estimativa f(x) = e-x em xi+1 = 1
para xi = 0,2. Use as versões de ordem zero, um, dois e três e calcule |εt| para
cada caso.

```python
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel
from math import exp
from numpy import linspace

f = lambda x: exp(-x)
df = lambda x: -exp(-x)

xini = 0.2
xfin = 1
delta = xfin - xini

plot(xini, f(xini), 'ro', label='x0') #plot(valor no eixo x, valor no eixo y, como é o pontinho, nome do pontinho na legenda)
plot(xfin, f(xfin), 'go', label='xi')

taylor0 = f(xini)
plot(xfin, taylor0, 'p', label='T0')

taylor1 = f(xini) + (df(xini)*(delta)) #fiz a funcao agr precisa marcar o ponto
plot(xfin, taylor1, 'p', label='T1')

taylor2 = taylor1 + (f(xini)*(delta**2)/2)
plot(xfin, taylor2, 'p', label='T2')

taylor3 = taylor2 + (df(xini)*(delta**3)/6)
plot(xfin, taylor3, 'p', label='T3')

# Erros 
real = f(xfin)
erro0 = abs((real - taylor0)/real)
erro1 = abs((real - taylor1)/real)
erro2 = abs((real - taylor2)/real)
erro3 = abs((real - taylor3)/real)

xs = linspace(0, 1.2, 200) #intervalo 0 ate 1, com 100 pontos
ys = [f(i) for i in xs]
plot(xs, ys)

grid() #serve pra botar o quadriculado no graf
legend() #cria a legenda com os labels
xlabel('x')
ylabel('f(x)') #nomeia os eixos

print("xi =", xini, " xf =", xfin, " h =", delta)
print("f(xi) =", f(xini))
print("f(xf) real =", real)
print("Taylor0 =", taylor0, "  |et| =", erro0)
print("Taylor1 =", taylor1, "  |et| =", erro1)
print("Taylor2 =", taylor2, "  |et| =", erro2)
print("Taylor3 =", taylor3, "  |et| =", erro3)
```

Use a expansão em série de Taylor de ordem zero até três para
prever f(3) para
f(x) = 25x3-6x2+7x-88
utilizando o ponto-base x=1. Calcule o erro relativo percentual εt para cada
aproximação.

```python
from matplotlib.pyplot import grid, plot, legend, xlabel, ylabel

f = lambda x: 25*x**3 - 6*x**2 + 7*x - 88
df = lambda x: 75*x**2 - 12*x + 7
df2 = lambda x: 150*x - 12
df3 = lambda x: 150

xi = 1
xf = 3
t = xf - xi

plot(xi, f(xi),'ro', label='x0')
plot(xf, f(xf), 'gd', label='xi')

t0 = f(xi)
plot(xf, t0, 'd', label='T0')

t1 = t0 + (df(xi)*t)
plot(xf, t1, 'd', label='T1')

t2 = t1 + (df2(xi)*(t**2)/2)
plot(xf, t2, 'd', label='T2')

t3 = t2 + (df3(xi)*(t**3)/6)
plot(xf, t3, 'd', label='T3')

#Erros
r = f(xf)
e0 = abs((r-t0)/r)*100
e1 = abs((r-t1)/r)*100
e2 = abs((r-t2)/r)*100
e3 = abs((r-t3)/r)*100

grid()
legend()
xlabel('x')
ylabel('f(x)')

print('xi=', xi, 'xf=', xf)
print("f(xi) =", f(xi))
print("f(xf) real =", r)
print("Taylor0 =", t0, "    |et| =", e0)
print("Taylor1 =", t1, "     |et| =", e1)
print("Taylor2 =", t2, "  |et| =", e2)
print("Taylor3 =", t3, "  |et| =", e3)
```

Use a expansão em série de Taylor de ordem zero até quatro para
prever f(2,5) para f(x) = ln x utilizando o ponto-base x =1. Calcule o erro relativo
percentual para cada aproximação. Discuta o significado dos resultados.

```python
from matplotlib.pyplot import grid, legend, xlabel, ylabel, plot
from numpy import log

f = lambda x: log(x)
df= lambda x: 1/x
df2= lambda x: -1/(x**2)
df3 = lambda x: 2/(x**3)
df4 = lambda x: -6/(x**4)
xinicial, xfinal = 1, 2.5
h = xfinal - xinicial
plot(xinicial, f(xinicial), 'rd', label='X0')
plot(xfinal, f(xfinal), 'rd', label='Xi') 

ty0 = f(xinicial)
plot(xfinal, ty0, 'o', label='T0')
ty1 = ty0 + (df(xinicial)*(h))
plot(xfinal, ty1, 'o', label='T1')
ty2 = ty1 + (df2(xinicial)*(h**2)/2)
plot(xfinal, ty2, 'o', label='T2')
ty3 = ty2 + (df3(xinicial)*(h**3)/6)
plot(xfinal, ty3, 'o', label='T3')
ty4 = ty3 + (df4(xinicial)*(h**4)/24)
plot(xfinal, ty4, 'o', label='T4')

#Erros
rl = f(xfinal)
er0 = abs((rl-ty0)/rl) * 100
er1 = abs((rl-ty1)/rl) * 100
er2 = abs((rl-ty2)/rl) * 100
er3 = abs((rl-ty3)/rl) * 100
er4 = abs((rl-ty4)/rl) * 100

grid(), legend(), xlabel('x'), ylabel('f(x)')

print('xinicial=', xinicial, 'xfinal= ', xfinal)
print("f(xi) =", f(xinicial), "\nf(xf) real =", rl)
print("Taylor0 =", round(ty0, 2), "    |et| =", er0)
print("Taylor1 =", round(ty1, 2), "     |et| =", er1)
print("Taylor2 =", round(ty2, 2), "  |et| =", er2)
print("Taylor3 =", round(ty3, 2), "  |et| =", er3)
print("Taylor4 =", round(ty4, 2), "  |et| =", er4)
```

A série de Taylor de ln(x) em torno de x=1 só converge para |x−1|<1. Como x=2.5 está fora desse intervalo, as aproximações não convergem para o valor real, podendo apresentar erros elevados e comportamento oscilatório.
