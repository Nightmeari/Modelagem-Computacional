```python
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel
from numpy import log

f=lambda x: log(x)
df = lambda x: 1/x
dff = lambda x: -1/x**2
df3 = lambda x: 2/x**3


xinicial = 1.5
xfinal = 2
plot(xinicial,f(xinicial),'ro',label='X0')
plot(xfinal,f(xfinal),'bd',label='Xi')

taylor1=f(xinicial)+(df(xinicial)*(xfinal-xinicial))
plot(xfinal,taylor1,'gd',label='Taylor ordem 1')

taylor2=taylor1 + (dff(xinicial)*((xfinal-xinicial)**2)/2)
plot(xfinal,taylor2,'p',label='Taylor ordem 2')

taylor3=taylor2 + (df3(xinicial)*((xfinal-xinicial)**3)/6)
plot(xfinal,taylor3,'o',label='Taylor ordem 3')

erro1 = (f(xfinal) - taylor1)/f(xfinal)
erro2 = (f(xfinal) - taylor2)/f(xfinal)
erro3 = (f(xfinal) - taylor3)/f(xfinal)

grid()
legend()
xlabel('x')
ylabel('f(x)')

print("f(xinicial) = ", f(xinicial))
print("f(xfinal) = ", f(xfinal))
print("Erros: ", erro1, " ", erro2, " ", erro3)
```

```python
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel, show
from numpy import log, linspace

f   = lambda x: log(x)
df  = lambda x: 1/x
d2f = lambda x: -1/x**2
d3f = lambda x: 2/x**3

x0 = 1.5
x1 = 2.0
h  = x1 - x0

# Taylor em torno de x0, avaliado em x1
t1 = f(x0) + df(x0)*h
t2 = f(x0) + df(x0)*h + d2f(x0)*h**2/2
t3 = f(x0) + df(x0)*h + d2f(x0)*h**2/2 + d3f(x0)*h**3/6

# Erros (absoluto e relativo)
fx1 = f(x1)
erro1_abs = abs(fx1 - t1)
erro2_abs = abs(fx1 - t2)
erro3_abs = abs(fx1 - t3)

erro1_rel = erro1_abs / fx1
erro2_rel = erro2_abs / fx1
erro3_rel = erro3_abs / fx1

# (Opcional) curva real para referncia
xs = linspace(1.2, 2.2, 200)
plot(xs, f(xs), label='f(x)=ln(x)')

# Pontos
plot(x0, f(x0), 'ro', label='x0')
plot(x1, fx1, 'bd', label='x1 (real)')

plot(x1, t1, marker='D', linestyle='None', label='Taylor ordem 1')
plot(x1, t2, marker='p', linestyle='None', label='Taylor ordem 2')
plot(x1, t3, marker='o', linestyle='None', label='Taylor ordem 3')

grid(True)
legend()
xlabel('x')
ylabel('f(x)')
show()

print("f(x0) =", f(x0))
print("f(x1) =", fx1)
print("Taylor 1 =", t1)
print("Taylor 2 =", t2)
print("Taylor 3 =", t3)
print("Erros abs:", erro1_abs, erro2_abs, erro3_abs)
print("Erros rel:", erro1_rel, erro2_rel, erro3_rel)
```
